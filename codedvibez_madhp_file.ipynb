````
import os
import zipfile

# --- 1. DEFINING THE FILE CONTENTS ---

html_content = """<!DOCTYPE html>
<html lang=\"en\">
<head>
  <meta charset=\"UTF-8\">
  <meta name=\"viewport\" content=\"width=device-width, initial-scale=1.0\">
  <title>CodedVibez | Pro Studio Dashboard</title>
  <link rel=\"stylesheet\" href=\"style.css\">
  <link rel=\"manifest\" href=\"manifest.json\">
  <meta name=\"theme-color\" content=\"#1e1e1e\">
</head>
<body>
  <header>
    <h1>Coded<span>Vibez</span></h1>
    <div style=\"font-size: 0.8rem; color: #888;\">Ultimate Studio v5.0</div>
  </header>
  <div class=\"container\">
    <div id=\"track\" class=\"panel\">
      <h2>Your Mix</h2>
      <canvas id=\"trackVisualizer\"></canvas>
      <input type=\"file\" id=\"trackUpload\" accept=\"audio/*\">
      <div class=\"vol-control\">
          <span>Gain</span>
          <input type=\"range\" id=\"vol-mix\" min=\"0\" max=\"1.5\" step=\"0.01\" value=\"1\">
      </div>
    </div>
    <div id=\"reference\" class=\"panel\">
      <h2>Reference</h2>
      <canvas id=\"refVisualizer\"></canvas>
      <input type=\"file\" id=\"referenceUpload\" accept=\"audio/*\">
      <div class=\"vol-control\">
          <span>Gain</span>
          <input type=\"range\" id=\"vol-ref\" min=\"0\" max=\"1.5\" step=\"0.01\" value=\"0.8\">
      </div>
    </div>
    <div id=\"ai\" class=\"panel\">
      <h2>AI Agent & Tools</h2>
      <div class=\"eq-buttons\">
          <button class=\"eq-btn\" id=\"btnLowCut\" onclick=\"toggleEQ('lowCut')\">‚úÇÔ∏è Cut Mud (200Hz)</button>
          <button class=\"eq-btn\" id=\"btnHighBoost\" onclick=\"toggleEQ('highBoost')\">‚ú® Air Boost (10kHz)</button>
          <button class=\"eq-btn\" id=\"btnMidScoop\" onclick=\"toggleEQ('midScoop')\">üé§ Vocal Pocket</button>
      </div>
      <hr style=\"border: 0; border-top: 1px solid #333; width: 100%; margin: 10px 0;\">
      <div id=\"ai-suggestions-container\">
        <ul id=\"aiSuggestions\">
          <li class=\"ai-msg\"><span class=\"ai-category ai-category-info\">Info</span> AI Ready. Load tracks to begin.</li>
        </ul>
      </div>
    </div>
    <div id=\"notepad\" class=\"panel\">
      <h2>Session Notes</h2>
      <textarea id=\"sessionNotes\" placeholder=\"Type lyrics or mix notes here... (Auto-saved)\"></textarea>
    </div>
    <div id=\"controls\">
      <button id=\"btn-play\" class=\"ctrl primary\" onclick=\"togglePlay()\">‚ñ∂ Play</button>
      <button class=\"ctrl\" onclick=\"stopTrack()\">‚èπ Stop</button>
      <button id=\"btn-ab\" class=\"ctrl\">‚öñ Hold Ref</button>
      <button id=\"btn-mono\" class=\"ctrl\" onclick=\"toggleMono()\">Mono</button>
      <button class=\"ctrl download\" onclick=\"downloadReport()\">‚¨á Report</button>
    </div>
  </div>
  <script src=\"script.js\"></script>
</body>
</html>"""

css_content = """:root {
  --bg-dark: #121212;
  --panel-bg: #1e1e1e;
  --accent: #00e676;
  --ref-accent: #ff4081;
  --text-main: #f0f0f0;
  --cat-eq: #66bb6a;
  --cat-dyn: #ffee58;
  --cat-stereo: #29b6f6;
  --cat-info: #bdbdbd;
}
body {
  font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
  background-color: var(--bg-dark);
  color: var(--text-main);
  margin: 0;
  height: 100vh;
  display: flex;
  flex-direction: column;
}
header {
  background-color: #000;
  padding: 15px 20px;
  display: flex;
  justify-content: space-between;
  align-items: center;
  border-bottom: 2px solid var(--accent);
}
h1 { margin: 0; font-size: 1.5rem; }
h1 span { color: var(--accent); }
.container {
  display: grid;
  flex: 1;
  grid-template-areas: \"track reference\" \"ai notepad\" \"controls controls\";
  grid-template-columns: 1fr 1fr;
  grid-template-rows: auto 1fr auto;
  grid-gap: 15px;
  padding: 15px;
  overflow: hidden;
}
.panel {
  background-color: var(--panel-bg);
  padding: 15px;
  border-radius: 12px;
  border: 1px solid #333;
  display: flex;
  flex-direction: column;
  position: relative;
}
#track { grid-area: track; border-top: 3px solid var(--accent); }
#reference { grid-area: reference; border-top: 3px solid var(--ref-accent); }
#ai { grid-area: ai; overflow-y: auto; }
#notepad { grid-area: notepad; display: flex; flex-direction: column; }
#controls { grid-area: controls; display: flex; align-items: center; justify-content: center; gap: 10px; background: #000; padding: 15px; }
h2 { margin: 0 0 10px 0; font-size: 1.1rem; color: #fff; }
canvas { background: #000; width: 100%; height: 100px; border-radius: 8px; margin-bottom: 10px; }
input[type=\"file\"] { margin-bottom: 10px; font-size: 0.8rem; background: #333; color: white; padding: 5px; border-radius: 4px; border: none; width: 100%; }
.vol-control { display: flex; align-items: center; gap: 10px; font-size: 0.9rem; }
input[type=\"range\"] { flex: 1; accent-color: var(--accent); cursor: pointer; }
.eq-buttons { display: flex; gap: 5px; margin-top: 10px; flex-wrap: wrap; }
.eq-btn { font-size: 0.8rem; padding: 5px 10px; background: #333; border: 1px solid #555; border-radius: 4px; color: #ccc; cursor: pointer; }
.eq-btn.active { background: var(--accent); color: #000; border-color: var(--accent); }
ul#aiSuggestions { list-style: none; padding: 0; margin: 0; display: flex; flex-direction: column; gap: 8px; }
li.ai-msg {
  background: #2a2a2a;
  padding: 8px 12px;
  border-radius: 6px;
  border-left: 3px solid var(--accent); /* Default highlight */
  font-size: 0.9rem;
  display: flex;
  align-items: flex-start;
  line-height: 1.4;
}
li.ai-alert { border-color: orange; color: #ffcc80; }

.ai-category {
  font-weight: bold;
  margin-right: 8px;
  padding: 2px 6px;
  border-radius: 4px;
  font-size: 0.75rem;
  text-transform: uppercase;
  color: #111;
}
.ai-category-eq { background-color: var(--cat-eq); }
.ai-category-dynamics { background-color: var(--cat-dyn); }
.ai-category-stereo { background-color: var(--cat-stereo); }
.ai-category-info { background-color: var(--cat-info); }

textarea { flex: 1; background: #111; color: #ddd; border: 1px solid #444; border-radius: 6px; padding: 10px; resize: none; font-family: monospace; }
button.ctrl { background: #333; color: white; border: none; padding: 12px 20px; border-radius: 6px; cursor: pointer; font-weight: bold; font-size: 1rem; }
button.ctrl:hover { background: #444; }
button.primary { background: var(--accent); color: #000; }
button.download { background: #2196f3; color: white; }
button.ab-active { background: var(--ref-accent); color: white; animation: pulse 1s infinite; }
button.mono-active { background: #ff9800; color: #000; }
@keyframes pulse { 0% { opacity: 1; } 50% { opacity: 0.7; } 100% { opacity: 1; } }
@media (max-width: 768px) {
    .container { grid-template-areas: \"track\" \"reference\" \"ai\" \"notepad\" \"controls\"; grid-template-columns: 1fr; grid-template-rows: auto; overflow-y: auto; }
    #notepad { height: 150px; }
}"""

js_content = """if ('serviceWorker' in navigator) {
    navigator.serviceWorker.register('sw.js').then(() => console.log('Service Worker Registered'));
}
const AudioContext = window.AudioContext || window.webkitAudioContext;
const ctx = new AudioContext();
const mixAnalyser = ctx.createAnalyser(); mixAnalyser.fftSize = 4096; mixAnalyser.smoothingTimeConstant = 0.85;
const refAnalyser = ctx.createAnalyser(); refAnalyser.fftSize = 4096; refAnalyser.smoothingTimeConstant = 0.85;
const mixGain = ctx.createGain();
const refGain = ctx.createGain();
const masterGain = ctx.createGain();
const lowCutFilter = ctx.createBiquadFilter(); lowCutFilter.type = 'highpass'; lowCutFilter.frequency.value = 200; lowCutFilter.Q.value = 0.7;
const highBoostFilter = ctx.createBiquadFilter(); highBoostFilter.type = 'highshelf'; highBoostFilter.frequency.value = 10000; highBoostFilter.gain.value = 0;
const midScoopFilter = ctx.createBiquadFilter(); midScoopFilter.type = 'peaking'; midScoopFilter.frequency.value = 500; midScoopFilter.gain.value = 0;
const splitter = ctx.createChannelSplitter(2);
const merger = ctx.createChannelMerger(2);
mixGain.connect(lowCutFilter);
lowCutFilter.connect(highBoostFilter);
highBoostFilter.connect(midScoopFilter);
midScoopFilter.connect(masterGain);
refGain.connect(refAnalyser);
refGain.connect(masterGain);
masterGain.connect(mixAnalyser);
let mixSource=null, refSource=null, mixBuffer=null, refBuffer=null, isPlaying=false, isMono=false, startTime=0, pauseTime=0, aiInterval=null;
document.getElementById('sessionNotes').value = localStorage.getItem('codedVibezNotes') || '';
document.getElementById('sessionNotes').addEventListener('input', (e) => localStorage.setItem('codedVibezNotes', e.target.value));
async function loadFile(file, isRef) {
  if (!file) return;
  addAiLog('Info', `‚è≥ Loading ${isRef ? 'Reference' : 'Mix'}...`);
  try {
    const arrayBuffer = await file.arrayBuffer();
    const buf = await ctx.decodeAudioData(arrayBuffer);
    if (isRef) { refBuffer = buf; addAiLog('Info', '‚úÖ Reference loaded. Ready.'); }
    else { mixBuffer = buf; addAiLog('Info', '‚úÖ Mix loaded. Ready.'); }
  } catch (e) { addAiLog('Error', '‚ùå Error: Could not load audio.'); }
}
document.getElementById('trackUpload').addEventListener('change', (e) => loadFile(e.target.files[0], false));
document.getElementById('referenceUpload').addEventListener('change', (e) => loadFile(e.target.files[0], true));
async function togglePlay() {
  if (ctx.state === 'suspended') await ctx.resume();
  if (isPlaying) {
    if (mixSource) { mixSource.stop(); mixSource = null; }
    if (refSource) { refSource.stop(); refSource = null; }
    pauseTime = ctx.currentTime - startTime;
    isPlaying = false;
    clearInterval(aiInterval);
    document.getElementById('btn-play').innerText = '‚ñ∂ Play';
  } else {
    if (!mixBuffer && !refBuffer) { addAiLog('Alert', "‚ö†Ô∏è No audio files loaded yet!"); return; }
    startTime = ctx.currentTime - pauseTime;
    if (mixBuffer) {
      mixSource = ctx.createBufferSource(); mixSource.buffer = mixBuffer; mixSource.connect(mixGain); mixSource.start(0, pauseTime);
      if(mixBuffer.duration < 10) mixSource.loop = true;
    }
    if (refBuffer) {
      refSource = ctx.createBufferSource(); refSource.buffer = refBuffer; refSource.connect(refGain); refSource.start(0, pauseTime); refGain.gain.value = 0;
    }
    connectOutput(); isPlaying = true; document.getElementById('btn-play').innerText = '‚è∏ Pause';
    drawVisualizer(); aiInterval = setInterval(runSmartAI, 3000);
  }
}
function stopTrack() { if (isPlaying) togglePlay(); pauseTime = 0; clearInterval(aiInterval); }
function connectOutput() {
  try { masterGain.disconnect(); splitter.disconnect(); } catch (e) {}
  if (isMono) {
    masterGain.connect(splitter); splitter.connect(merger, 0, 0); splitter.connect(merger, 0, 1); splitter.connect(merger, 1, 0); splitter.connect(merger, 1, 1); merger.connect(ctx.destination);
  } else { masterGain.connect(ctx.destination); }
}
function toggleMono() {
  isMono = !isMono;
  const btn = document.getElementById('btn-mono');
  if (isMono) { btn.classList.add('mono-active'); addAiLog('Stereo', '‚ö†Ô∏è Output is MONO.'); } else { btn.classList.remove('mono-active'); }
  if (isPlaying) connectOutput();
}
function getBandAverage(analyser, startFreq, endFreq) {
    const bufferLength = analyser.frequencyBinCount;
    const dataArray = new Uint8Array(bufferLength);
    analyser.getByteFrequencyData(dataArray);
    const sampleRate = ctx.sampleRate;
    const startBin = Math.floor(startFreq / (sampleRate / 2) * bufferLength);
    const endBin = Math.floor(endFreq / (sampleRate / 2) * bufferLength);
    let sum = 0, count = 0;
    for (let i = startBin; i < endBin && i < bufferLength; i++) { sum += dataArray[i]; count++; }
    return count === 0 ? 0 : sum / count;
}

// New helper for overall loudness proxy (average energy across all frequency bins)
function getOverallEnergy(analyser) {
    const bufferLength = analyser.frequencyBinCount;
    const dataArray = new Uint8Array(bufferLength);
    analyser.getByteFrequencyData(dataArray);
    let sum = 0;
    for (let i = 0; i < bufferLength; i++) {
        sum += dataArray[i];
    }
    return sum / bufferLength; // Average energy across all bands
}

// New helper for peak detection proxy (maximum amplitude in frequency domain)
function getMaxPeakFrequencyDomain(analyser) {
    const bufferLength = analyser.frequencyBinCount;
    const dataArray = new Uint8Array(bufferLength);
    analyser.getByteFrequencyData(dataArray);
    let maxVal = 0;
    for (let i = 0; i < bufferLength; i++) {
        if (dataArray[i] > maxVal) {
            maxVal = dataArray[i];
        }
    }
    return maxVal;
}

function runSmartAI() {
    if (!isPlaying || !mixBuffer) return;

    // --- Frequency Band Analysis ---
    const bands = {
        sub: { min: 20, max: 60 },
        bass: { min: 60, max: 250 },
        lowMid: { min: 250, max: 500 },
        mid: { min: 500, max: 2000 },
        highMid: { min: 2000, max: 6000 },
        air: { min: 6000, max: 16000 }
    };
    // More specific bands for detailed EQ suggestions
    const specificEqBands = {
        muddyFreq: { min: 250, max: 350 }, // Common muddy area
        boxyFreq: { min: 350, max: 500 },  // Common boxy area
        presenceFreq: { min: 3000, max: 5000 }, // Vocal presence
        harshFreq: { min: 4000, max: 8000 } // Sibilance/harshness
    };

    const mixLevels = {};
    for (const [name, r] of Object.entries(bands)) mixLevels[name] = getBandAverage(mixAnalyser, r.min, r.max);
    for (const [name, r] of Object.entries(specificEqBands)) mixLevels[name] = getBandAverage(mixAnalyser, r.min, r.max);

    let refLevels = null;
    if (refBuffer) {
        refLevels = {};
        for (const [name, r] of Object.entries(bands)) refLevels[name] = getBandAverage(refAnalyser, r.min, r.max);
        for (const [name, r] of Object.entries(specificEqBands)) refLevels[name] = getBandAverage(refAnalyser, r.min, r.max);
    }

    // --- Dynamic Analysis ---
    const mixOverallEnergy = getOverallEnergy(mixAnalyser);
    const mixMaxPeak = getMaxPeakFrequencyDomain(mixAnalyser); // Proxy for peak in freq domain

    let refOverallEnergy = 0;
    let refMaxPeak = 0;
    if (refBuffer) {
        refOverallEnergy = getOverallEnergy(refAnalyser);
        refMaxPeak = getMaxPeakFrequencyDomain(refAnalyser);
    }

    // --- AI Suggestions ---

    // EQ Suggestions
    if (refLevels) {
        // Comparative EQ suggestions
        const mixBalance = mixLevels.bass / (mixLevels.highMid + 1); // Simple proxy for bass-treble balance
        const refBalance = refLevels.bass / (refLevels.highMid + 1);

        if (mixBalance > refBalance * 1.3 && mixLevels.lowMid > refLevels.lowMid * 1.2) {
            addAiLog('EQ', "üìâ Mix sounds muddy compared to reference. Consider a subtle cut around 250-350Hz.");
        } else if (mixBalance < refBalance * 0.7 && mixLevels.bass < refLevels.bass * 0.8) {
            addAiLog('EQ', "üìà Mix sounds thin compared to reference. Try boosting bass around 80-120Hz.");
        }

        if (mixLevels.presenceFreq < refLevels.presenceFreq * 0.8) {
            addAiLog('EQ', "üé§ Vocals might be buried. Consider boosting presence around 3-5kHz.");
        }
        if (mixLevels.harshFreq > refLevels.harshFreq * 1.2) {
            addAiLog('EQ', "üëÇ Highs might be harsh or sibilant. Try a gentle cut around 4-8kHz.");
        }
        if (mixLevels.lowMid > refLevels.lowMid * 1.2 && mixLevels.boxyFreq > refLevels.boxyFreq * 1.2) {
            addAiLog('EQ', "üì¶ Mix sounds boxy compared to reference. Check 350-500Hz.");
        }

    } else {
        // General EQ suggestions without reference
        if (mixLevels.lowMid > mixLevels.highMid * 1.3 && mixLevels.muddyFreq > 100) {
            addAiLog('EQ', "‚ö†Ô∏è Mud detected (250-350Hz). Consider a low-mid cut.");
        }
        if (mixLevels.sub > mixLevels.bass * 1.2 && mixLevels.sub > 180) {
            addAiLog('EQ', "üîä Sub is too loud. Consider taming sub-bass around 20-60Hz.");
        }
        if (mixLevels.harshFreq > 120) {
            addAiLog('EQ', "üò¨ Mix might be too bright or harsh. Listen to 4-8kHz.");
        }
        if (mixLevels.presenceFreq < 50 && mixLevels.mid < 60) {
             addAiLog('EQ', "üéß Mix sounds dull. A gentle boost in upper-mids (3-5kHz) might help.");
        }
    }

    // Dynamics Suggestions
    if (refBuffer) {
        if (mixOverallEnergy < refOverallEnergy * 0.7) {
            addAiLog('Dynamics', "üìâ Mix overall loudness is low compared to reference. Consider compression or gain staging.");
        } else if (mixOverallEnergy > refOverallEnergy * 1.2) {
            addAiLog('Dynamics', "üìà Mix might be too loud or over-compressed compared to reference.");
        }

        // Simple peak-to-loudness ratio proxy for dynamic range
        const mixDynamicRangeProxy = mixMaxPeak / (mixOverallEnergy + 1);
        const refDynamicRangeProxy = refMaxPeak / (refOverallEnergy + 1);

        if (mixDynamicRangeProxy > refDynamicRangeProxy * 1.5 && mixOverallEnergy > 50) {
            addAiLog('Dynamics', "üéöÔ∏è Mix dynamic range seems wide. Consider gentle compression to even things out.");
        } else if (mixDynamicRangeProxy < refDynamicRangeProxy * 0.5 && mixOverallEnergy > 150) {
            addAiLog('Dynamics', "üß± Mix sounds squashed. Check for over-compression or limiting.");
        }
    } else {
        // General dynamics suggestions without reference
        if (mixMaxPeak > 240 && mixOverallEnergy > 180) { // arbitrary high values for potential issues
            addAiLog('Dynamics', "üö® Potential clipping or harsh peaks detected. Check levels or apply a limiter.");
        } else if (mixOverallEnergy < 50) {
            addAiLog('Dynamics', "üí° Mix sounds quiet. Increase gain or apply subtle compression.");
        }
    }

    // Stereo Imaging Suggestions (Very basic, based on mono state)
    // Note: True stereo imaging analysis requires separate L/R channel analysers, which is beyond the scope of this single-analyser setup.
    // A simplified check: if masterGain is outputting to a single channel (mono) or if our mono toggle is active.
    const isCurrentlyMonoOutput = masterGain.channelCount === 1 || isMono; // masterGain's channelCount reflects the active connection
    if (isCurrentlyMonoOutput) {
        addAiLog('Stereo', "üßê Currently in Mono. Remember to check your mix in Stereo too!");
    } else {
        // If we had L/R correlation data, we'd put specific suggestions here.
        // For now, a general reminder.
        addAiLog('Stereo', "‚ÜîÔ∏è Ensure your stereo image has good width and depth. Pan elements for clarity.");
    }
}

function addAiLog(category, text, isAlert = false) {
  const ul = document.getElementById('aiSuggestions');
  // Check if the exact message (including category) is already the first child to avoid duplicates
  if (ul.firstChild && ul.firstChild.innerText.includes(text)) return;

  const li = document.createElement('li');
  li.className = 'ai-msg' + (isAlert ? ' ai-alert' : '');

  const categorySpan = document.createElement('span');
  categorySpan.className = `ai-category ai-category-${category.toLowerCase()}`;
  categorySpan.innerText = category;

  li.appendChild(categorySpan);
  li.appendChild(document.createTextNode(text));

  ul.prepend(li);
  // Keep the list to a reasonable length (e.g., last 8 messages)
  if (ul.children.length > 8) ul.lastChild.remove();
}
function toggleEQ(type) {
  if (type === 'lowCut') {
    const btn = document.getElementById('btnLowCut');
    if (lowCutFilter.type === 'highpass') { lowCutFilter.frequency.value = 0; lowCutFilter.type = 'allpass'; btn.classList.remove('active'); } // Disables filter
    else { lowCutFilter.type = 'highpass'; lowCutFilter.frequency.value = 200; btn.classList.add('active'); addAiLog('EQ', 'Applied: Low Cut (200Hz).'); } // Enables filter
  }
  if (type === 'highBoost') {
    const btn = document.getElementById('btnHighBoost');
    if (highBoostFilter.gain.value === 0) { highBoostFilter.gain.value = 5; btn.classList.add('active'); addAiLog('EQ', 'Applied: +5dB Air Boost (10kHz).'); } // Enables boost
    else { highBoostFilter.gain.value = 0; btn.classList.remove('active'); } // Disables boost
  }
  if (type === 'midScoop') {
    const btn = document.getElementById('btnMidScoop');
    if (midScoopFilter.gain.value === 0) { midScoopFilter.gain.value = -4; btn.classList.add('active'); addAiLog('EQ', 'Applied: -4dB Mid Scoop (500Hz).'); } // Enables scoop
    else { midScoopFilter.gain.value = 0; btn.classList.remove('active'); } // Disables scoop
  }
}
function downloadReport() {
    const notes = document.getElementById('sessionNotes').value;
    const aiLogs = Array.from(document.querySelectorAll('#aiSuggestions li')).map(li => `- ${li.innerText}`).join('\n');
    const content = `CODEDVIBEZ REPORT\nDate: ${new Date().toLocaleString()}\n\n-- NOTES --\n${notes}\n\n-- AI ANALYSIS --\n${aiLogs}`;
    const blob = new Blob([content], { type: 'text/plain' });
    const url = URL.createObjectURL(blob);
    const a = document.createElement('a'); a.href = url; a.download = `CodedVibez_Report.txt`; a.click();
}
const btnAB = document.getElementById('btn-ab'), volMix = document.getElementById('vol-mix'), volRef = document.getElementById('vol-ref');
const startRef = () => { if (!refBuffer) return; mixGain.gain.value = 0; refGain.gain.value = volRef.value; btnAB.classList.add('ab-active'); btnAB.innerText = 'Listening to Ref...'; };
const endRef = () => { mixGain.gain.value = volMix.value; refGain.gain.value = 0; btnAB.classList.remove('ab-active'); btnAB.innerText = '‚öñ Hold Ref'; };
btnAB.addEventListener('mousedown', startRef); btnAB.addEventListener('mouseup', endRef);
btnAB.addEventListener('touchstart', (e) => { e.preventDefault(); startRef(); }); btnAB.addEventListener('touchend', (e) => { e.preventDefault(); endRef(); });
volMix.addEventListener('input', (e) => { if (!btnAB.classList.contains('ab-active')) mixGain.gain.value = e.target.value; });
function drawVisualizer() {
  if (!isPlaying) return;
  requestAnimationFrame(drawVisualizer);
  const bufferLength = mixAnalyser.frequencyBinCount;
  const dataArray = new Uint8Array(bufferLength);
  mixAnalyser.getByteFrequencyData(dataArray);
  const canvas = document.getElementById('trackVisualizer'); const c = canvas.getContext('2d');
  c.fillStyle = '#1e1e1e'; c.fillRect(0, 0, canvas.width, canvas.height);
  const barWidth = (canvas.width / bufferLength) * 2.5; let x = 0;
  for (let i = 0; i < bufferLength; i++) {
    const barHeight = dataArray[i] / 2.5; c.fillStyle = `rgb(${barHeight + 100}, 230, 118)`; c.fillRect(x, canvas.height - barHeight, barWidth, barHeight); x += barWidth + 1;
  }
  if (refBuffer) {
    const refCanvas = document.getElementById('refVisualizer'); const rc = refCanvas.getContext('2d');
    const rData = new Uint8Array(refAnalyser.frequencyBinCount); refAnalyser.getByteFrequencyData(rData);
    rc.fillStyle = '#1e1e1e'; rc.fillRect(0, 0, refCanvas.width, refCanvas.height);
    let rx = 0; for (let i = 0; i < bufferLength; i++) { const h = rData[i] / 2.5; rc.fillStyle = `rgb(255, ${h + 50}, 150)`; rc.fillRect(rx, refCanvas.height - h, barWidth, h); rx += barWidth + 1; }
  }
}"""

manifest_content = """{
  \"name\": \"CodedVibez Studio\",
  \"short_name\": \"CodedVibez\",
  \"start_url\": \"index.html\",
  \"display\": \"standalone\",
  \"background_color\": \"#1e1e1e\",
  \"theme_color\": \"#00e676\",
  \"icons\": [
    {
      \"src\": \"https://cdn-icons-png.flaticon.com/512/3075/3075977.png\",
      \"sizes\": \"192x192\",
      \"type\": \"image/png\"
    }
  ]
}"""

sw_content = """self.addEventListener('install', (e) => {
    e.waitUntil(
      caches.open('codedvibez-store').then((cache) => cache.addAll([
        'index.html',
        'style.css',
        'script.js',
        'manifest.json',
      ])),
    );
  });
self.addEventListener('fetch', (e) => {
    e.respondWith(
      caches.match(e.request).then((response) => response || fetch(e.request)),
    );
  });"""

# --- 2. WRITING THE FILES ---

files = {
    "index.html": html_content,
    "style.css": css_content,
    "script.js": js_content,
    "manifest.json": manifest_content,
    "sw.js": sw_content
}

zip_filename = "CodedVibez_Studio.zip"

print("Creating CodedVibez Studio files...")

with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for filename, content in files.items():
        # Write individual files (optional, but good for local use)
        with open(filename, 'w', encoding='utf-8') as f:
            f.write(content)
        # Add to zip
        zipf.write(filename)
        print(f"Packed: {filename}")

print(f"\n‚úÖ SUCCESS! '{zip_filename}' has been created in this folder.")
````
